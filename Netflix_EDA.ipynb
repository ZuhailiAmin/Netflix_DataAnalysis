## Step 1: Install and Import Relevant Libraries and Packages

We will require NumPy, Pandas, Matplotlib and Seaborn, so I will be installing and importing them.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load and Explore the Dataset

Let's start by loading the dataset and taking an initial look at the first few rows.

```python
netflix_titles=pd.read_csv("netflix_titles.csv")
netflix_titles.head()
```

Checking for missing values:

```python
netflix_titles.isnull().sum()
```

Handling missing values:

```python
netflix_titles.dropna()
```

## Step 3: Data Preprocessing

Converting 'date_added' to datetime and adding new columns for month and year:

```python
netflix_titles["date_added"] = pd.to_datetime(netflix_titles['date_added'])

netflix_titles['month_added']=netflix_titles['date_added'].dt.month
netflix_titles['month_name_added']=netflix_titles['date_added'].dt.month_name()
netflix_titles['year_added'] = netflix_titles['date_added'].dt.year

netflix_titles.head()
```

## Step 4: Finding a Specific Director

Let's find rows with a specific director, e.g., 'Kirsten Johnson'.

```python
specific_director_rows = netflix_titles[netflix_titles['director'] == 'Kirsten Johnson']
specific_director_rows
```

## Step 5: Data Visualization

Count of Shows by Type (Bar chart):

```python
plt.figure(figsize=(12, 10))
sns.countplot(x='type', data=netflix_titles, palette='Set2')
plt.title('Count of Shows by Type')
plt.show()
```

Count of Shows by Type (Pie chart):

```python
content_distribution = netflix_titles['type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(content_distribution, labels=content_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Distribution of Content Types on Netflix')
plt.show()
```

Title Count by Country

```python
titles_by_country = netflix_titles.groupby('country')['title'].count().reset_index()
titles_by_country.sort_values(by='title', ascending=False)
```

```python
top_countries = netflix_titles['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with the Most Titles on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

Boxplot of Release Years by Show Type:

```python
plt.figure(figsize=(12, 10))
sns.boxplot(x='type', y='release_year', data=netflix_titles, palette='muted')
plt.title('Boxplot of Release Years by Show Type')
plt.show()
```

Movies Duration Distribution:

```python
netflix_movies_df = netflix_titles[netflix_titles['type'] == 'Movie'].copy()
netflix_movies_df['duration'] = netflix_movies_df['duration'].str.replace(' min', '')
netflix_movies_df['duration'] = pd.to_numeric(netflix_movies_df['duration'], errors='coerce')
netflix_movies_df['duration']
```

```python
sns.set(style="darkgrid")
sns.kdeplot(data=netflix_movies_df['duration'], shade=True)
```
